```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
countries=pd.read_csv("world-data-2023.csv")
countries.head()
```

```python
countries.tail()
```

### Data Cleaning
Data was cleaned in excel as we researched the missing values individually. We felt that we should not assume what the data is as it is based on factual country data and we wanted to try and answer question without skewed data, so instead of using mean or median we did the research. The data that we filled in was pulled from the world bank's data collections. 

```python
countries.info()
```

## Removing unwanted data

Columns we plan on removing: 'Abbreviation', 'Armered_Forces_size','Calling_Code', 'Capital/Major_City', 'Largest_city', 'Out_of_pocket_health_expenditure','Forested_Area_(%)','Tax_revenue_(%)','Latitude', 'Longitude'

We removed these columns as they seemed didn't relate to the questions we wanted to examine or weren't particularly relevant to the dataset overall. 

### Replacing the " " between words in column titles with _ to allow for use with Pandas

```python
countries.columns =[column.replace(" ", "_") for column in countries.columns]
```

```python
countries.head()
```

### Removing Columns
The columns we decided to move were Country abbreviation, calling code, capital/major city, and largest city. We removed these as they aren't really relevant to our data analysis. 

```python
countries.drop(['Abbreviation', 'Armed_Forces_size', 'Calling_Code', 'Capital/Major_City', 'Largest_city', 'Out_of_pocket_health_expenditure','Forested_Area_(%)','Tax_revenue_(%)','Latitude', 'Longitude'], 
        axis=1, 
        inplace=True)
```

```python
countries.info()
```
